Problem 1: Social Media – Post Engagement Analyzer

```python
#fun to analyze
def analyze(likes):

    total_likes = 0

    #Calculate total likes using loop
    for like in likes:
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


#Example data
likes = [100, 100, 400, 250]

analyze(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
#function to analyze rainfall

def analyze_rainfall(rainfall):
#initilize total to 0
    total = 0

    # total rainfall using for loop
    for value in rainfall:
        total += value

    average = total / len(rainfall)

    print("Average Rainfall:", int(average))#flroat to int conversion

    required_level = 70   #Minimum required rainfall level

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


#Example rainfall data
rainfall = [10, 15, 80, 55, 79]

analyze_rainfall(rainfall)
```

Problem 4: Social Media – Duplicate Account Detection

```python
#function to check duplicate account
def check_duplicates(usernames):

    unique_users = set(usernames)   #Remove duplicates using set

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


#Example usernames
users = ["anita", "subhash", "anoop", "smith","subhash"]

#function call
check_duplicates(users)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
#function for filtering crops
def premium_crops(prices):
#intilize empty list
    premium_list = []

    #Loop through each price
    for price in prices:
        if price > 2000:
            premium_list.append(price)

    return premium_list


#Example crop prices
crop_prices = [1800, 2020, 2026, 2700, 1700]

#Call the function
crops = premium_crops(crop_prices)

print("Premium Crops:", crops)
```

Problem 7: System Monitoring – Application Health

```python
#function to health check
def evaluate_health(error_count):
    print(f"Error count: {error_count}")
    if error_count == 0:
        return "Healthy"
    elif error_count <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"


#input from user 
errors = int(input("Error Count: "))

#function call
status = evaluate_health(errors)

print("System Status:", status)
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
#function for transaction
def validate_transaction(amount):
#set limit
    daily_limit = 50000
    print(f"Transaction Amount: {amount}")
    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"


#input from user
transaction_amount = int(input("Transaction Amount: "))

#function call
status = validate_transaction(transaction_amount)

print("Transaction Status:", status)
```

Problem 9: E-Learning – Student Attendance Eligibility

```python
#function for eligibility
def check_eligibility(attendance):
#set count 0
    present_count = 0

    #Count present days using loop
    for day in attendance:
        if day == "P":
            present_count += 1

    total_days = len(attendance)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


#Example attendance data
attendance = ["A", "P", "A", "A", "P","P"]

#function call
check_eligibility(attendance)
```

Problem 10: Smart Electricity Bill Analyzer

```python
#function for bill
def calculate_bill(units):
#first set to 0
    bill = 0
    print(f"total units consumed: {units}")
    #Rfor first 100 units
    if units <= 100:
        bill = units * 3

    #for next 100 units till 200
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    #for above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
#return both values in tuple

    return bill, status


#input from user
units = int(input("Enter units consumed: "))

#function call
total, usage_status = calculate_bill(units)

print("Total Bill: ₹", total)
print("Usage Level:", usage_status)
```
