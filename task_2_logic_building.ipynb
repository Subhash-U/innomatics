Problem 1: Unique Words in a Sentence

```python
#input from user
sentence = input("Enter sentence: ")

# Splitting sentence into separate words
words = sentence.split()

# Convert list to set to get unqiue elements
unique_words = set(words)

# Display
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
# Dict to stor employee salary data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize variables to track maximum salary
max_sal = 0
max_sal_name = ""

# Iterating through dictionary
for name, salary in employees.items():
    if salary > max_sal:
        max_sal = salary
        max_sal_name = name

# Display 
print("Highest Salary:", max_sal_name, "-", max_sal)
```

Problem 3: Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element of list
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display
print("list:", numbers)
print("max:", maximum)
print("min:", minimum)
```

Problem 4: Count Products Above a Price Threshold


```python
prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop through price list
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# Count Present days
for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
attendance_per = (present_count / len(attendance)) * 100


print("Attendance Percentage:", attendance_per)
```

Problem 7: Count Character Frequency

```python
text = "python"

char_frequency = {}

# Count character occurrences
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1
#display
print(char_frequency)
```

Problem 10: Calculate Average Marks



```python
marks = [70, 75, 80, 65]

total = 0

# Sum of  marks
for mark in marks:
    total += mark

# Calculate average
average = total / len(marks)

#display
print("Average Marks:", average)
```
