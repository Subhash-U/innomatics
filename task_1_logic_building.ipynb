```python
#1. User Login Check

#initilize username and password

username = "admin"
password = "1234"
#asking user to Enter

username_a = input("Enter username:")
password_b = input("Enter password:")

#conditional flow

if username_a == username and password_b == password:
    print('Login Successful')
else:
    print('Invalid Credentials')
```

```python
#2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

#for loop
for mark in marks:
    #conditional flow
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
#print statements
print("Total Students:", len(marks))
print("Total Passed:", pass_count)
print("Total Failed:", fail_count)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []#list
for name in names:
    cleaned = name.strip().lower()#converting
    cleaned_names.append(cleaned)
print("cleaned names:", cleaned_names)
```

```python
#4. Message Length Analyzer

#input message
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)#give length 
    print("Message:", msg)
    print("Length:", length)
    #conditional flow
    if length > 10:
        print("Status: Long Message!")
    else:
        print("Status: Short Message")
    
```

```python
#5. Error Message Detector
#input list
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")#count function 
print("Total ERROR entries (count):", error_count)#print the function value
```
