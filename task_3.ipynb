1: Employee Performance Bonus Eligibility

```python
#input in dict of name:score
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find the highest score
highest_score = max(employees.values())

# Find all employees who have the highest score ,intilize empty list
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

2: Search Query Keyword Analysis


```python
#input the string
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
import string
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

3: Sensor Data Validation


```python
#input sensor data in lsit

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Store hour_index, reading_value for even readings

#intilize empty list
valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

 4: Email Domain Usage Analysis


```python
#input of email in list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

#dict intialize
domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # Extract domain part after '@'
    domain_count[domain] = domain_count.get(domain, 0) + 1

# display percentage 
total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

 5: Sales Spike Detection



```python
#input of sales in list
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Determine spike 30% above average
threshold = average_sales * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

 6: Duplicate User ID Detection


```python
#input of id
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences of user ID
frequency = {}

for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1

# Display if more than 1 time
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
