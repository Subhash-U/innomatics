Problem 1: Mobile Recharge Validation System


```python
#function for recharge checking
def check_recharge(recharge_value):

    #list of plans
    available_plans = [199, 299, 399, 599]
    if recharge_value < 50:
        return "Minimum recharge amount is 50"
    if recharge_value not in available_plans:
        return "Selected recharge plan is not available"
    return "OK"

#indefintite loop

while True:
    amount = int(input("Enter recharge amount: "))
    validation_status = check_recharge(amount)
    if validation_status == "OK":
        print(f"Recharge  Successful for {amount}!")
        break
    else:
        print("Recharge Failed:", validation_status)
        print("Please try again.\n")
```

Problem 2: Inventory Reorder Alert System


```python
#function
def inventory(stock_data):
    for product in stock_data:
        available_units = stock_data[product]
        if available_units < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")
#dict for records

warehouse_records = {
    "Soap": 12,
    "brush": 25,
    "Shampoo": 8,
    "water bottles": 30,
    "juices":3
}
#func call
inventory(warehouse_records)
```

Problem 3: Student Result Processing System


```python
#func for student result

def student_result(mark_list):
    #set score to 0
    total_score = 0
    for score in mark_list:
        total_score = total_score + score
    sub_count = len(mark_list)
    avg_score = total_score / sub_count
    if avg_score >= 50:
        return "Pass"
    else:
        return "Fail"

#input of marks as list

student_marks = [60, 45, 70, 55]
final_result = student_result(student_marks)
print("Result:", final_result)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
#fuction
def trip_cost(distance_km, peak_status):
    #set base standard price
    base_charge = 50
    #charge per km
    per_km_charge = 12

    #toatal
    total_amt = base_charge + (distance_km * per_km_charge)

    #check if peak hours
    if peak_status == "yes":
        total_amt = total_amt + (0.25 * total_amt)
    return total_amt
#infinte loop
while True:
    travel_distance = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()
    final_fare = trip_cost(travel_distance, peak_input)
    print("Estimated Fare: ₹", final_fare)
    retry_choice = input("Do you want to calculate again? (yes/no): ").lower()
    if retry_choice != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python
#function to check employee attendence
def check_attendance(attendance_records):

    present_counter = 0
#for loop for traverse
    for day_mark in attendance_records:
        if day_mark == "P":
            present_counter += 1

    total_days_count = len(attendance_records)
    attendance_percentage = (present_counter / total_days_count) * 100

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

#input for func
employee_log = ["P", "P", "A", "P", "P"]

final_result = check_attendance(employee_log)

print("Attendance Status:", final_result)
```

Problem 6: Password Strength Checker


```python
#function to check password strength
def validate_password_strength(user_password):

    has_digit_flag = False
    has_special_flag = False
    allowed_special = "@#$"

    if len(user_password) < 8:
        return "Weak Password (Minimum length 8 required)"

    for character_item in user_password:
        if character_item.isdigit():
            has_digit_flag = True

        if character_item in allowed_special:
            has_special_flag = True

    if has_digit_flag and has_special_flag:
        return "Strong Password"
    else:
        return "Weak Password (Missing digit or special character)"

#hardcoded test password
test_password = "abc@1212"

result = validate_password_strength(test_password)

print(result)
```
